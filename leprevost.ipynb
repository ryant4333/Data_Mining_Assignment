```python
A = df[df['gname']=='Shining Path (SL)']
sns.distplot(A['iyear'],kde=False,bins=47)
plt.title('SL acts over the years', fontsize=18)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
```
